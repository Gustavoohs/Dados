```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cota = pd.read_csv('C:/Users/Pichau/Desktop/cotas.csv')
cota['Data'] = pd.to_datetime(cota['Data'], format='%d/%m/%Y')
```

```python
cota['mmv'] = cota['Media'].rolling(30, min_periods=1).mean()
cota['mmv2'] = cota['Media'].rolling(60, min_periods=1).mean()
```

```python
plt.figure(figsize=(15,8))

plt.plot(cota['Data'], cota['Media'], linewidth=2, alpha=0.5)
plt.plot(cota['Data'], cota['mmv'], color='green', linewidth=3, alpha=0.7)
plt.plot(cota['Data'], cota['mmv2'], color='red', linewidth=3)

plt.text(cota['Data'][48], 102184,
        "Mínima histórica\n(5,3% volume útil)\nnovembro de 2017", fontsize=10)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['Cota mensal', 'Média móvel 2,5 anos', 'Média móvel 5 anos'],
          fontsize=14)
plt.title("Reservatório Descoberto", fontsize=16)
plt.xlabel('Anos',fontsize=16)
plt.ylabel('Cota [cm]', fontsize=16)
plt.show()
```
