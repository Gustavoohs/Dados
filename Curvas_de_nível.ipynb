```python
# Bibliotecas
from osgeo import gdal
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Lendo imagem como array
img = gdal.Open("DEM.tif")
array = img.ReadAsArray().astype(np.float)
```

```python
# Visualizando curvas de nível com matplotlib
fig = plt.figure(figsize = (12, 8))
plt.contour(np.flip(array, axis=0), cmap = "viridis", 
            levels = list(range(array.min().astype(int), 
                                array.max().astype(int), 10)))
plt.colorbar()
plt.show()
```
