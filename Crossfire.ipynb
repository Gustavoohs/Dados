```python
# Bibliotecas
import geopandas as gpd
from crossfire import AsyncClient
from crossfire import occurrences
from shapely.geometry import Point
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
# Baixando dados com asyncio
client = AsyncClient(email="email", password="senha")
await client.states()
await client.cities(city_name='Recife')
df = await client.occurrences('813ca36b-91e3-4a18-b408-60b27a1942ef', format='df')
```

```python
# Criar uma geometria de ponto usando Shapely
geometry = [Point(xy) for xy in zip(df['longitude'].astype(float), df['latitude'].astype(float))]

# Criar um GeoDataFrame
geo_df = gpd.GeoDataFrame(df, geometry=geometry)
```

```python
#Extrairas coordenadas para o basemap
x_map=geo_df.centroid.x.mean()
y_map=geo_df.centroid.y.mean()

#Criar o Basemap
m = folium.Map(location=[y_map, x_map], zoom_start=10)
```

```python
# List comprehension para as coordenadas
heat_data = [[row['latitude'],row['longitude']] for index, row in geo_df.iterrows()]
```

```python
# Visualizar mapa de calor
HeatMap(heat_data, alpha=0.01).add_to(m)
m
```
