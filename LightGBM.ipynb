```python
# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
# Lendo imagem como array
src = rio.open('L71221071_07120010720_DN.tif')
b1 = src.read(1)
b2 = src.read(2)
b3 = src.read(3)
b4 = src.read(4)
b5 = src.read(5)
b7 = src.read(6)

img = np.dstack([b1,b2,b3,b4,b5,b7])
```

```python
# Lendo classes
with rio.open('/content/roi.tif') as src2:
  roi = src2.read(1)
```

```python
# Criando conjunto de treino
X = img[roi > 0]
y = roi[roi > 0]
```

```python
# Dividindo conjunto em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y.ravel(), 
                                                    test_size = 0.1)
```

```python
# Instanciando classificador
LGBM = lgb.LGBMClassifier()
LGBM.fit(X_train, y_train)
```

```python
# Criando matriz com bandas em colunas
img2d = img.reshape(img.shape[0]*img.shape[1], img.shape[2])

# Rodando predição
pred_lgbm = LGBM.predict(img2d)

# Retornando para o formato original
pred_lgbm_final = pred_lgbm.reshape(img[:,:,0].shape)
```

```python
# Visualizando resultado
plt.imshow(pred_lgbm_final, cmap='hsv_r')
```
