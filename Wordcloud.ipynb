```python
# Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# Lendo arquivo
with open('dados.txt') as f:
    strings = f.readlines()
```

```python
# Concatenando todas as palavras
palavras = " ".join(s for s in strings)
```

```python
# Definindo stopwords
palavras = palavras.lower()
stopwords = set(STOPWORDS)
stopwords.update(['um','que','em','não','para','os','são','de','da','como',
                  'por','mais','na','dos','se','isso','uma','das','mas',
                  'está','já','ao'])
```

```python
# Lendo imagem para máscara
mask = np.array(Image.open("/content/Design sem nome.png"))
   
# Criando nuvem
wordcloud = WordCloud(stopwords=stopwords,colormap='jet',
                      background_color='black', 
                      width=3000, height=3000, max_words=15000,
                      mask=mask, max_font_size=200,
                      min_font_size=1).generate(palavras)

# Visualizando resultado
fig, ax = plt.subplots(figsize=(15,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud)
```
