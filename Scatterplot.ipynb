```python
import tifffile as tiff
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
b2 = tiff.imread('B2.tif')
b3 = tiff.imread('B3.tif')               
```

```python
b2 = b2.flatten().reshape(-1,1)
b3 = b3.flatten().reshape(-1,1)
```

```python
reg = linear_model.LinearRegression()
modelo = reg.fit(b2,b3)
reta = modelo.predict(b2)
r2 = modelo.score(b2,b3)
```

```python
plt.style.use("ggplot")
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(b2, reta, color='r')
ax.scatter(b2, b3, edgecolor='k', facecolor='blue', alpha=0.7)
ax.set_title('$R^2= %f$' % r2, fontsize=18)
ax.set_ylabel('Banda 3', fontsize=14)
ax.set_xlabel('Banda 2', fontsize=14)

fig.tight_layout()
```
