```python
# Bibliotecas
import geopandas as gpd
import folium
from folium.plugins import HeatMap
import geopandas as gpd
```

```python
# Lendo arquivo
est = gpd.read_file('estacoes.shp')
```

```python
# Extraindo coordenadas
loc = list(zip(est.geometry.y, est.geometry.x))

# Criando pontos Folium baseado no geodataframe
pts = folium.features.GeoJson(est)

# Criando visualização base
m = folium.Map(location=[-15.788497,-47.879873],
               zoom_start=10, control_scale=True)

# Criando mapa de calor e adicionando ao mapa base
# parametros: HeatMap(data, name=None, min_opacity=0.5, max_zoom=18, max_val=1.0, 
#radius=25, blur=15, gradient=None, overlay=True, control=True, show=True)
HeatMap(loc).add_to(m)
pts.add_to(m)
m
```

```python
# Salvando mapa em html
m.save('teste.html')
```
