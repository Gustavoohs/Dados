```python
# Biblioteca
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
# Lendo arquivo original
gdf = gpd.read_file('/content/poligono_teste.shp')
```

```python
# Visualizando polígono com basemap
fig, ax = plt.subplots(figsize=(12,8))

gdf.plot(ax=ax, linewidth =  0.03, alpha=0.3)

ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, crs=gdf.crs)
```

```python
# Convertendo para formato GeoPackage
gdf.to_file('poligono_teste.gpkg', driver='GPKG')
```

```python
# Convertendo para formato Geojson
gdf.to_file('poligono_teste.geojson', driver='GeoJSON')
```
